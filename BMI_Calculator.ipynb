Project Question:

BMI Calculator with Loop Function
Write a Python program to create a BMI (Body Mass Index) calculator that continuously prompts the user to input their weight and height until they choose to exit the program. The program should calculate and display the BMI based on the following formula:
BMI=weight (kg)height (m)2\text{BMI} = \frac{\text{weight (kg)}}{\text{height (m)}^2}BMI=height (m)2weight (kg)

Requirements:
1.	Input Handling:
o	Prompt the user to enter their weight in kilograms (kg) and height in meters (m).
o	Include error handling to ensure that the inputs are valid numbers and are greater than zero.

2.	BMI Calculation:
o	Calculate the BMI using the provided weight and height.

3.	BMI Classification:
o	Display the BMI value and classify it into one of the following categories:
	Underweight: BMI < 18.5
	Normal weight: 18.5 ≤ BMI < 24.9
	Overweight: 25 ≤ BMI < 29.9
	Obesity: BMI ≥ 30

4.	Looping:
o	After displaying the BMI and its classification, ask the user if they want to calculate another BMI.
o	If the user inputs "yes," prompt them to enter new weight and height.
o	If the user inputs "no," exit the program and display a farewell message.

5.	Exit Condition:
o	Ensure the loop continues until the user decides to exit by typing "no."



```python
#Normal BMI = (18.5 - 24.9)kg/m2

weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in m: "))

BMI = weight/(height**2)

if BMI < 18.5:
    print ("Underweight")

elif BMI <= 24.9:
    print("Normal weight")

elif BMI <= 29.9:
    print ("Overweight")

else:
    print ("Obese")




```

```python
def calculate_bmi():
    """Function to calculate and display BMI"""
    try:
        # Input handling
        weight = float(input("Enter your weight in kilograms: "))
        height = float(input("Enter your height in meters: "))

        # Validate inputs
        if weight <= 0 or height <= 0:
            print("Error: Weight and height must be greater than zero.")
            return False

        # BMI Calculation
        bmi = weight / (height ** 2)

        # Display BMI value
        print(f"\nYour BMI is: {bmi:.2f}")

        # BMI Classification
        if bmi < 18.5:
            print("Classification: Underweight")
        elif 18.5 <= bmi < 24.9:
            print("Classification: Normal weight")
        elif 25 <= bmi < 29.9:
            print("Classification: Overweight")
        else:
            print("Classification: Obesity")

        return True

    except ValueError:
        print("Error: Please enter valid numbers for weight and height.")
        return False

# Main loop using for loop
for _ in range(10000):  # Large number to simulate an infinite loop
    if calculate_bmi():
        # Ask if the user wants to calculate another BMI
        for _ in range(3):  # Give the user 3 attempts to enter a valid response
            choice = input("\nDo you want to calculate another BMI? (yes/no): ").lower()
            if choice in ['yes', 'no']:
                break
            print("Invalid input. Please enter 'yes' or 'no'.")
        
        # Exit condition
        if choice != 'yes':
            print("Thank you for using the BMI calculator. Goodbye!")
            break
    else:
        print("Let's try again.")
```
